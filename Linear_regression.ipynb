SECTION2 Linear Regression

```python
def hypothesis(theta,x):
    y_=theta[0]+theta[1]*x
    return y_


def gradient(X,Y,theta):
    grads=np.zeros((2,))
    m=X.shape[0]
    for i in range(m):
        Y_=hypothesis(theta,X[i])
        y=Y[i]
        grads[0]+=(Y_-y)
        grads[1]+=(Y_-y)*X[i]
    return grads/m


def error(X,Y,theta):
    m=X.shape[0]
    err=0
    for i in range(m):
        Y_=hypothesis(theta,X[i])
        y=Y[i]
        err+=(Y_-y)**2
    return err/m


def gradient_descent(X,Y,learning_rate=0.1):
    theta=np.zeros((2,))
    theta_list=[]
    error_list=[]
    for i in range(100):
        grad=gradient(X,Y,theta)
        e=error(X,Y,theta)
        
        
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
    return theta,error_list,theta_list     
        
```

```python
theta,error_list,theta_list=gradient_descent(x_train,y_train)
```
