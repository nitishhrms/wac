```python
def distance(X1,X2):
    return np.sqrt(((X1-X2)**2).sum(axis=0))
```

```python
distance(np.array([1,2]),np.array([3,4]))
```

```python
def knn(x,y,query_point,k):
    ##no of examples
    m=x.shape[0]
    ans=[]
    for i in range(m):
        d=distance(query_point,x[i])
        
        ans.append((d,y[i]))
        
        ans1=sorted(ans)[:k]
        ans2=np.array(ans1)
        ans3=np.unique(ans2[:,1],return_counts=True)
        index=np.argmax(ans3[1])
    return ans3[0][index]
        
```
